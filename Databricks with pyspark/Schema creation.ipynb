#Importing dataset and Creating schema for it

```python
from pyspark.sql.types import StructField,StructType, StringType, IntegerType, DoubleType , FloatType ,DateType
abs_path="/Volumes/workspace/default/sampledataset/jsonfiles/grocery_chain_data.json"
data_schema=[
StructField('customer_id', StringType(), True),
StructField('store_name', StringType(), True),
StructField('transaction_date', DateType(), True),
StructField('aisle', StringType(), True),
StructField('product_name', StringType(), True),
StructField('quantity', StringType(), True),
StructField('unit_price', FloatType(), True),
StructField('total_amount', FloatType(), True),
StructField('discount_amount', FloatType(), True),
StructField('final_amount', FloatType(), True),
StructField('loyalty_points', IntegerType(), True),
]
final_struc=StructType(fields=data_schema)
df=spark.read.json(abs_path,schema=final_struc).limit(20)
#because dataset was  long we set limit to 20 for simple tasks
df.display()
df.printSchema()
print(df.columns)
df.show()
```

#Withcolumn attribute
>used to add new column or update an existing column


```python
df.withColumn('after_tax_price', df['unit_price'] * 1.1).show()
```

```python
df.withColumn('quantity', df['quantity']+ 2.0).show()
```

#Sql queries in pyspark
>we register a dataframe as a temporary sql table which allows us to write sql queries directly

```python
df.createOrReplaceTempView('Grocery')
result=spark.sql("select product_name,quantity from Grocery")
result.show()
```

#groupby attribute

```python
df.groupBy('store_name').avg('unit_price').show()
```

```python
from pyspark.sql.functions import count
df.groupBy('product_name').agg(count('product_name').alias('sales')).show()
```

```python
from pyspark.sql.functions import min,max,select
df.groupBy('product_name').agg(max('unit_price').alias('maximum price among stores')).show()
```

```python
df.groupBy('product_name').agg(min('unit_price').alias('minimum price among stores')).show()
```

```python
df.select(avg('discount_amount').alias('average discount')).show()
```

```python
from pyspark.sql.functions import countDistinct,avg,stddev
df.select(countDistinct('product_name').alias('total variety of products')).show()
```

#standard deviation

```python
price_std=df.select(stddev('unit_price').alias('standard deviation')).show()
```
